It would be better to change the time variables from ints to categorical variables, and change time variables to the datetime dtype.

```python
train[["placed_time", "confirmed_time", "arrive_pickup_time", "pickup_time", "delivered_time"]] = train[["placed_time", "confirmed_time", "arrive_pickup_time", "pickup_time", "delivered_time"]].apply(pd.to_datetime, format = "%I:%M:%S %p")
```

```python
train.dtypes # checking the new datetime dtypes
```

```python
# changing time-related int dtype columns to categorical variables.
for column in train.columns:
    if train[column].dtype == "int64":
        train[column] = train[column].astype("object")
```

```python
# change distance_km and time_pickup_to_delivered back to int dtype.
train = train.astype({"distance_km": "int64", "time_pickup_to_arrival": "int64"})
```

```python
train.dtypes # checking the new categorical or object dtypes
```

Check for missing values

Notice that the Temprature and the Precipitation columns have missing values. Since the missing values in the precipitation columns is more than 90%, we need to drop that column

```python
import missingno as msno
msno.matrix(train)
```

```python
train.describe()
```

```python
# dropping the columns with the missing values
train.drop(columns="precipitation_ml", inplace=True)
```

```python
# imputing the temprature col
train.temp.fillna(train.temp.mean(), inplace=True)
```

```python
train.describe()
```

```python
# let's join riders to train.
riders = pd.read_csv("Riders.csv")
riders.head()
```

```python
train = pd.merge(train, riders, on = "Rider Id", how = "left")
```

```python
train.head() # we now have 32 columns
```

#### Conclusion 
- There are no duplicate rows in the data.
- The variables temp and precipitation_ml have missing values. precipitation_ml has been removed from the dataset. temp was imputed with its mean.
- All original int variables were changed to object, with the exception for distance_km and time_pickup_to_arrival, which have to be int.
- All 5 time variables were changed to datetime dtype instead of object.

```python
train.to_csv("train2.csv", index=False)
```
