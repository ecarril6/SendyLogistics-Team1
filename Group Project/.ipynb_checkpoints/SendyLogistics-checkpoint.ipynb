```python
train = pd.read_pickle("data/sendy3.pkl")
train.head()
```

```python
train.columns
```

```python
# Dropping date/time columns
drop_columns = ['placed_day', 'placed_wkday', 'placed_time', 'confirmed_day',
       'confirmed_wkday', 'confirmed_time', 'arrive_pickup_day',
       'arrive_pickup_wkday', 'arrive_pickup_time', 'pickup_day',
       'pickup_wkday', 'pickup_time', 'delivered_day', 'delivered_wkday',
       'delivered_time']

train.drop(columns=drop_columns, inplace=True)
```

```python
train.head()
```

```python
# Creating dummy variable for personal_or_business (1 if personal 0 if business)

train["personal"] = 0
train.loc[train.customer_type == "Personal", "personal"] = 1
```

```python
# Creating dummy variables for platforms 1, 2, 3. 

def create_platform_dummies(df):
    for i in range(1, 4):
        plat_str = "platform" + str(i)
        df[plat_str] = 0
        df.loc[df.platform_type == i, plat_str] = 1
create_platform_dummies(train)
```

```python
drop_col= ['order_no', 'user_id', 'platform_type', 'customer_type']
train.drop(columns=drop_col, inplace=True)
```
