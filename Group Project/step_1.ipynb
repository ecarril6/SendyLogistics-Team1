## The Goal
- To predict the estimated time of delivery of orders, from the point of driver pickup to the point of arrival at final destination.

## The Data Dictionary
- order_no (object) : Unique number identifying the order
- user_id (object) : Unique number identifying the customer on a platform
- vehicle_type (object) : For this competition limited to bikes, however in practice Sendy service extends to trucks and vans
- platform_type (int64) : Platform used to place the order, there are 4 types
- customer_type (object) : Customer type
- plac_day_of_month (int64) : Placement - Day of Month i.e 1-31
- plac_weekday (int64) : Placement - Weekday (Monday = 1)
- plac_time (object) : Placement - Time - Time of day the order was placed
- conf_day_of_month (int64) : Confirmation - Day of Month i.e 1-31
- conf_weekday (int64) : Confirmation - Weekday (Monday = 1)
- conf_time (object) : Confirmation - Time - Time of day the order was confirmed by a rider
- arrival_pickup_day_of_month (int64) : Arrival at Pickup - Day of Month i.e 1-31
- arrival_pickup_weekday (int64) : Arrival at Pickup - Weekday (Monday = 1)
- arrival_pickup_time (object) : Time of day the the rider arrived at the location to pick up the order - as marked by the rider through the Sendy application
- pickup_day_of_month (int64) : Pickup - Day of Month i.e 1-31
- pickup_weekday (int64) : Pickup - Weekday (Monday = 1)
- pickup_time (object) : Pickup - Time - Time of day the the rider picked up the order - as marked by the rider through the Sendy application
- arrival_des_day_of_month (int64): Arrival at Delivery - Day of Month i.e 1-31
- arrival_des_weekday (int64) : Arrival at Delivery - Weekday (Monday = 1)
- arrival_des_time (object) : Arrival at Delivery Time - Time of day the rider arrived at the destination to deliver the order - as marked by the rider through the Sendy application
- distance_km (int64) : The distance from Pickup to Destination
- Temperature (float64) : Temperature at the time of order placement in Degrees Celsius (measured every three hours)
- precipitation (float64) : Precipitation at the time of order placement (measured every three hours)
- pickup_lat (float64) : Latitude of pick up location
- pickup_long (float64) : Longitude of pick up location
- destination_lat (float64) : Latitude of delivery location
- destination_long (float64) : Longitude of delivery location
- rider_id (object) : ID of the Rider who accepted the order
- time_from_pickup_to_arrival (int64) : Time in seconds between 'Pickup' and 'Arrival at Destination' - calculated from the columns for the purpose of facilitating the task
